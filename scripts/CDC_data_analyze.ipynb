# Clean

```python
# Create df copy to safely manipulate
df2 = df.copy()
```

```python
# Remove "Missing" ages
df2 = df2[df2['age_group'] != "Missing"]
df2["age_group"].value_counts()
```

```python
# Remove "sex" entries with small counts
df2 = df2[df2["sex"] != "Other"]
df2 = df2[df2["sex"] != "Missing"]
df2 = df2[df2["sex"] != "Unknown"]
df2["sex"].value_counts()
```

```python
# Remove select "race" entries
df2 = df2[df2["race"] != "Unknown"]
df2 = df2[df2["race"] != "Missing"]
df2 = df2[df2["race"] != "Multiple/Other"]
df2["race"].value_counts()
```

```python
# Convert inconclusive ICU entries to "No"
df2["icu_yn"].replace("Missing","No",inplace=True)
df2["icu_yn"].replace("Unknown","No",inplace=True)
df2["icu_yn"].value_counts()
```

```python
# Convert inconclusive hospitalization entries to "No"
df2["hosp_yn"].replace("Missing","No",inplace=True)
df2["hosp_yn"].replace("Unknown","No",inplace=True)
df2["hosp_yn"].value_counts()
```

```python
# Convert inconclusive death entries to "No"

df2["death_yn"].replace("Missing","No",inplace=True)
df2["death_yn"].replace("Unknown","No",inplace=True)
df2["death_yn"].value_counts()
```

```python
# Remove columns
df2.drop(columns=['process','ethnicity','state_fips_code','county_fips_code','exposure_yn','symptom_status'],inplace=True)
df2.head()
```

```python
df_count = df['case_month'].count()
df2_count = df2['case_month'].count()
df_diff = df_count - df2_count

print(f'Original DF length: {df_count:,.2f}')
print(f'Cleaned DF length: {df2_count:,.2f}')
print(f'Difference: {df_diff:,.2f}')
```
